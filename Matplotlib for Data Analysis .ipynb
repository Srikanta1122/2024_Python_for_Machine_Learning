```python
plt.figure(figsize=(10,6))
plt.subplot(1,3,1)
plt.plot(data['Size'], data['Juiciness'], color='r',linestyle='--',linewidth=2, marker='o',markersize=5, markerfacecolor='y', 
        markeredgecolor='b', markeredgewidth=3, label='Hii')
plt.subplot(1,3,2) # 1st pos is row, 2nd pos is columns, 3rd pos is position
plt.hist(data['Size']) 
plt.subplot(1,3,3)
plt.scatter(data['Size'], data['Juiciness'])
```
