```python
sns.catplot(x='day', y='total_bill', data=data, jitter=False)
```

```python
#bar plot
titanic=sns.load_dataset('titanic')
```

```python
sns.catplot(x='sex', y='survived', kind='bar', data=titanic, hue='class') 
```
